```r
rawdata<-read.csv("network_backup_dataset.csv")
```

```r
mydata<-data.frame(rawdata)
```

```r
colnames(mydata)
```

```r
Numeric.mydata = transform( mydata[, c(1,2,3,4,5,6,7)], Work.Flow.ID = unclass(Work.Flow.ID), File.Name = unclass(File.Name) )
# Numeric.mydata[which(Numeric.mydata$Day.of.Week=='Monday'),]<-'1'
Numeric.mydata$Day.of.Week <- ifelse(Numeric.mydata$Day.of.Week=='Monday',1,ifelse(Numeric.mydata$Day.of.Week=='Tuesday',2,ifelse(Numeric.mydata$Day.of.Week=='Wednesday',3,ifelse(Numeric.mydata$Day.of.Week=='Thursday',4,ifelse(Numeric.mydata$Day.of.Week=='Friday',5,ifelse(Numeric.mydata$Day.of.Week=='Saturday',6,ifelse(Numeric.mydata$Day.of.Week=='Sunday',7,0)))))))
flow0file<-unique(Numeric.mydata[which(Numeric.mydata$Work.Flow.ID==1),]$File.Name)
flow1file<-unique(Numeric.mydata[which(Numeric.mydata$Work.Flow.ID==2),]$File.Name)
flow2file<-unique(Numeric.mydata[which(Numeric.mydata$Work.Flow.ID==3),]$File.Name)
flow3file<-unique(Numeric.mydata[which(Numeric.mydata$Work.Flow.ID==4),]$File.Name)
```

```r
workflow0<-Numeric.mydata[which(Numeric.mydata$Work.Flow.ID==1 & Numeric.mydata$File.Name==flow0file),]
day = c((workflow0$Week..-1)*7+workflow0$Day.of.Week)
sizePerDay = rep(0,max(day))
for(i in 1:length(day)){
    sizePerDay[(workflow0[i,]$Week..-1)*7+workflow0[i,]$Day.of.Week] = sizePerDay[(workflow0[i,]$Week..-1)*7+workflow0[i,]$Day.of.Week]+workflow0[i,]$Size.of.Backup..GB.
}

plot(sizePerDay, main="WorkFlow0", ylab="Backup Size(GB)", xlab="Day")
hour = c((workflow0$Week..-1)*7*(workflow0$Day.of.Week-1)+workflow0$Backup.Start.Time...Hour.of.Day)
sizePerHour = rep(0,max(hour))
for(i in 1:length(hour)){
    sizePerHour[(workflow0[i,]$Week..-1)*7*(workflow0[i,]$Day.of.Week-1)+workflow0[i,]$Backup.Start.Time...Hour.of.Day] = sizePerHour[(workflow0[i,]$Week..-1)*7*(workflow0[i,]$Day.of.Week-1)+workflow0[i,]$Backup.Start.Time...Hour.of.Day]+workflow0[i,]$Size.of.Backup..GB.
}
plot(sizePerHour, main="WorkFlow0", ylab="Backup Size(GB)", xlab="Hour")
```

```r
workflow1<-Numeric.mydata[which(Numeric.mydata$Work.Flow.ID==2 & Numeric.mydata$File.Name==flow1file),]
day = c((workflow1$Week..-1)*7+workflow1$Day.of.Week)
sizePerDay = rep(0,max(day))
for(i in 1:length(day)){
    sizePerDay[(workflow1[i,]$Week..-1)*7+workflow1[i,]$Day.of.Week] = sizePerDay[(workflow1[i,]$Week..-1)*7+workflow1[i,]$Day.of.Week]+workflow1[i,]$Size.of.Backup..GB.
}

plot(sizePerDay, main="WorkFlow0", ylab="Backup Size(GB)", xlab="Day")
hour = c((workflow1$Week..-1)*7*(workflow1$Day.of.Week-1)+workflow1$Backup.Start.Time...Hour.of.Day)
sizePerHour = rep(0,max(hour))
for(i in 1:length(hour)){
    sizePerHour[(workflow1[i,]$Week..-1)*7*(workflow1[i,]$Day.of.Week-1)+workflow1[i,]$Backup.Start.Time...Hour.of.Day] = sizePerHour[(workflow1[i,]$Week..-1)*7*(workflow1[i,]$Day.of.Week-1)+workflow1[i,]$Backup.Start.Time...Hour.of.Day]+workflow1[i,]$Size.of.Backup..GB.
}
plot(sizePerHour, main="WorkFlow0", ylab="Backup Size(GB)", xlab="Hour")
```

```r
k.folds <- function(k) {
    folds <- createFolds(Numeric.mydata$Size.of.Backup..GB., k = k, list = TRUE, returnTrain = TRUE)
    for (i in 1:k) {
        model <- lm(Size.of.Backup..GB. ~ ., data=Numeric.mydata[folds[[i]],])
        predictions <- predict(object=model, newdata=Numeric.mydata[-folds[[i]],])
        all_predictions <- c(all_predictions, predictions)
        err <- c(err, rmse(predictions,Numeric.mydata[-folds[[i]],]$Size.of.Backup..GB.))
        print(summary(model))
    }
#     all_predictions
    err
}

set.seed(100)
err <- c()
# all_predictions <- c()
# all_predictions <- k.folds(10)
# all_predictions
err <- k.folds(10)
err
folds <- createFolds(Numeric.mydata$Size.of.Backup..GB., k = 10, list = TRUE, returnTrain = TRUE)
model <- lm(Size.of.Backup..GB. ~ ., data=Numeric.mydata[folds[[7]],])
testset <- Numeric.mydata[-folds[[7]],]
predictions <- predict(object=model, newdata=testset)

# library(ggplot2)
df_lm <- data.frame(predicted = predictions, observed = testset$Size.of.Backup..GB.)
ggplot(df_lm, aes(x = predicted, y = observed)) +
  geom_point() +
  geom_abline(intercept = 0, slope = 1, colour = "blue") +
  ggtitle("Linear Regression Predicted VS Observed")

df_lm2 <- data.frame(predicted = predictions, residual = testset$Size.of.Backup..GB. - predictions)
ggplot(df_lm2, aes(x = predicted, y = residual)) +
  geom_point() +
  geom_hline(yintercept = 0, colour = "blue") +
  ggtitle("Linear Regression Predicted VS Residual")
```
